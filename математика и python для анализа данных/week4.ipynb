```python
import numpy as np
import pandas as pd
from scipy.stats import expon
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
```

```python
#create a set of 1000 random variables from exponential distribution
rvs = expon.rvs(size=1000)

#create argument from 0 to 10
x = np.linspace(0, 5, 1000)

#plot histogram of rvs
plt.hist(rvs, normed=True)

#plot theoretical probability density function (pdf)
plt.plot(x, expon.pdf(x), 'r-', lw=5)

plt.ylabel('$f(x)$')
plt.xlabel('$x$')
```

```python
n_5 = []
for i in range(1, 1001):
    n_5.append(np.random.choice(rvs, 5))
    
n_15 = []
for i in range(1, 1001):
    n_15.append(np.random.choice(rvs, 15))    

n_30 = []
for i in range(1, 1001):
    n_30.append(np.random.choice(rvs, 30))
```

```python
means_for_n_5 = []
for i in n_5:
    means_for_n_5.append(i.mean())
    
means_for_n_15 = []
for i in n_15:
    means_for_n_15.append(i.mean())
    
means_for_n_30 = []
for i in n_30:
    means_for_n_30.append(i.mean())
```

```python
plt.hist(means_for_n_5, facecolor='b', normed=True)
plt.ylabel('fraction of samples')
plt.xlabel('$x$')
```

```python
plt.hist(means_for_n_15, facecolor='m', normed=True)
plt.ylabel('fraction of samples')
plt.xlabel('$x$')
```

```python
plt.hist(means_for_n_30, facecolor='w', normed=True)
plt.ylabel('fraction of samples')
plt.xlabel('$x$')
```

```python
plt.hist(means_for_n_5, facecolor='b', normed=True)
plt.hist(means_for_n_15, facecolor='m', normed=True)
plt.hist(means_for_n_30, facecolor='w', normed=True)

plt.ylabel('fraction of samples')
plt.xlabel('$x$')
```

```python
#theoretical mean and variance
m, v = expon.stats(moments='mv')
```

```python
#parameters of normal distribution
norm_rv_n_5 = sts.norm(m, (v/5)**(1/2))
norm_rv_n_15 = sts.norm(m, (v/15)**(1/2))
norm_rv_n_30 = sts.norm(m, (v/30)**(1/2))

x_1 = np.linspace(-2,2,1000)
pdf_norm_5 = norm_rv_n_5.pdf(x_1)
pdf_norm_15 = norm_rv_n_15.pdf(x_1)
pdf_norm_30 = norm_rv_n_30.pdf(x_1)

plt.plot(x_1, pdf_norm_5,'b--', lw=2, label='theoretical pdf (n=5)', alpha=0.5)
plt.plot(x_1, pdf_norm_15,'m--', lw=2, label='theoretical pdf (n=15)', alpha=0.5)
plt.plot(x_1, pdf_norm_30,'g--', lw=2, label='theoretical pdf (n=30)', alpha=0.5)

plt.hist(means_for_n_5, facecolor='b', normed=True)
plt.hist(means_for_n_15, facecolor='m', normed=True)
plt.hist(means_for_n_30, facecolor='w', normed=True)

plt.legend()
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
```
